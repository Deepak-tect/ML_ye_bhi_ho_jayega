### Transformation

```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'PRODUCT_CONDITION')
```

```python
plot_data(df,'missing')
```
